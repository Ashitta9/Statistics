a. The company wishes to clarify whether there is any
increase in sales after stepping into digital marketing


T Test
H0:  no significant difference
H1: significant increase.

```python

before = data2['Sales_before_digital_add(in $)']
after = data2['Sales_After_digital_add(in $)']
```

```python
from scipy.stats import ttest_ind
before = data2['Sales_before_digital_add(in $)']
after = data2['Sales_After_digital_add(in $)']
t_stat1,p_value=ttest_ind(before,after)
print (t_stat1)
print(p_value)
```

```python
if p_value <0.05:
    print("Reject the null hypothesis: There is a significant increase in sales.")
else:
    print("Fail to reject the null hypothesis: No significant increase in sales.")
```

b. The company needs to check whether there is any
dependency between the features “Region” and
“Manager”.


CHiSquare test

H0 : Region and Manager  are independent
H1:Region and Manager  are dependent.

```python
from scipy.stats import chisquare
from scipy.stats import chi2_contingency
```

```python
observed = pd.crosstab(data2['Region'],data2['Manager'])
```

```python
print(observed)
```

```python

chi2_stat, p_value1, dof,exp = chi2_contingency(observed)
print(chi2_stat)
print(p_value1)
```

```python
if p_value1 < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
